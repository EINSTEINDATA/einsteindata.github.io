# One Variable (categorical)

```python
# Product Family Desc sales
print(df_sales['Product Family Desc'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# sales by Product Family Desc')
plt.xlabel('2017')
plt.ylabel('Product Family Desc')

labels = df_sales['Product Family Desc'].value_counts().index
values = df_sales['Product Family Desc'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels, rotation='vertical')
plt.show()
```
